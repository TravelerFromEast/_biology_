```python
# translation
aa_codon_dictionary = {
    "A": ["GCU", "GCC", "GCA", "GCG"],
    "R": ["CGU", "CGC", "CGA", "CGG", "AGA", "AGG"],
    "N": ["AAU", "AAC"],
    "D": ["GAU", "GAC"],
    "B": ["AAU", "AAC", "GAU", "GAC"],
    "C": ["UGU", "UGC"],
    "Q": ["CAA", "CAG"],
    "E": ["GAA", "GAG"],
    "Z": ["CAA", "CAG", "GAA", "GAG"],
    "G": ["GGU", "GGC", "GGA", "GGG"],
    "H": ["CAU", "CAC"],
    "I": ["AUU", "AUC", "AUA"],
    "L": ["CUU", "CUC", "CUA", "CUG", "UUA", "UUG"],
    "M": ["AUG"], 
    "K": ["AAA", "AAG"], 
    "F": ["UUU", "UUC"],
    "P": ["CCU", "CCC", "CCA", "CCG"],
    "S": ["UCU", "UCC", "UCA", "UCG", "AGU", "AGC"],
    "T": ["ACU", "ACC", "ACA", "ACG"],
    "W": ["UGG"],
    "Y": ["UAU", "UAC"],
    "V": ["GUU", "GUC", "GUA", "GUG"],
    "--STOP--": ["UAA", "UGA", "UAG"]
}


def valid_strand(strand):
    flag = False
    for x in strand:
        if x not in ["A", "T", "G", "C"]:
            flag = True
    return not flag
    

def complement(strand):
    complement_strand = ""
    for x in strand:
        if x == "A":
            complement_strand+= "T"
        elif x == "T":
            complement_strand+= "A"
        elif x == "G":
            complement_strand+= "C"
        else:
            complement_strand+= "G"
    return complement_strand


def transcribe(strand):
    strand = complement(strand)
    return strand.replace("T", "U")


def check_mRNA_length(strand):
    return len(strand) >= 3


def translate(template_strand):
    mRNA = transcribe(template_strand)
    protein = ""
    
    for i in range(0, len(mRNA) - 2, 3):
        codon = mRNA[i:i+3]
        for aa, codons in aa_codon_dictionary.items():
            if codon in codons:
                if aa == "--STOP--":
                    return protein
                protein += aa
                break
    return protein



dna = input("Enter template DNA strand: ").strip().upper()

if valid_strand(dna):
    mRNA = transcribe(dna)
    print("mRNA:", mRNA)
    
    if check_mRNA_length(mRNA):
        protein = translate(dna)
        print("Protein:", protein)
    else:
        print("Length of mRNA too short to be translated.")
else:
    print("Invalid DNA template strand.")
```

```python
# gc content
def valid_strand(strand):
    flag = False
    for x in strand:
        if x not in ["A", "T", "G", "C"]:
            flag = True
            break
    return not flag


def complement(strand):
    complementary_strand = ""
    n_dictionary = {"A" : "T", "T" : "A", "G" : "C", "C" : "G"}
    for x in strand:
        for base, complement_base in n_dictionary.items():
            if x == base:
                complementary_strand += complement_base
                break
    return complementary_strand


def gc_fraction(strand):
    count = 0
    combined_strand = strand + complement(strand)
    for x in combined_strand:
        if x in ["G", "C"]:
            count += 1
    return count / (2 * len(strand))


def at_gc_fraction(strand):
    count_1, count_2 = 0, 0
    combined_strand = strand + complement(strand)
    
    for x in combined_strand:
        if x in ["C", "G"]:
            count_1 += 1
        else:
            count_2 += 1
    return count_1/count_2


dna = input("enter a strand: ").strip().upper()

if valid_strand(dna) == True:
    print("GC content: ", gc_fraction(dna), end="\n") 
    print("AT-CG fraction: ", at_gc_fraction(dna), end="\n")

else:
    print("input strand invalid")
```

# HANDLING AMBIGUITY

*N:* Any nucleotide (A, T, G, or C).
*R:* Purine (A or G).
*Y:* Pyrimidine (C or T).
*S:* Strong (G or C).
*W:* Weak (A or T).
*K:* Keto (G or T).
*M:* Amino (A or C).
*B:* Not A (C, G, or T).
*D:* Not C (A, G, or T).
*H:* Not G (A, C, or T).
*V:* Not T (A, C, or G).
For RNA, replace T with U (e.g., Y = C or U).

```python
# ATGNGTRCG
# complement and reverse complement
def complement(strand):
    n_dictionary = {"A" : "T", "T" : "A", "G" : "C", "C" : "G", "N" : "N", "R" : "Y", "Y" : "R", "S" : "W", "W" : "S", "M": "K", "K" : "M"}
    complementary_strand = ""
    for x in strand:
        for base, complement_base in n_dictionary.items():
            if x == base:
                complementary_strand += complement_base
                break
    return complementary_strand


def reverse_complement(strand):
    strand = complement(strand)[::-1]
    return strand


def ambiguity(strand):
    flag = False
    for x in strand:
        if x in ["N", "R", "Y", "S", "W", "K", "M"]:
            flag = True
            break
    return flag
    

def valid_strand(strand):
    flag = False
    for x in strand:
        if x not in ["A", "T", "G", "C", "N", "R", "Y", "S", "W", "K", "M"]:
            flag = True
            break
    return not flag


dna = input("enter strand: ").strip().upper()

if valid_strand(dna) == True:
    if ambiguity(dna) == True:
        print("ambiguity detected", end="\n")
    else:
        print("ambiguity absent", end="\n")
    print("DNA Strand: ", dna, end="\n")
    print("Complementary Strand: ", complement(dna), end="\n")   
    print("Reverse Complement: ", reverse_complement(dna), end="\n") 
else:
    print("input strand invalid")
```
