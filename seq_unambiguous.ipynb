```python
# create complementary strand: 
# DNA strand: TACGCATGC
def valid_strand(strand):
    flag = False
    for x in strand: 
        if x not in ["A", "T", "G", "C"]:
            flag = True
            break
    return not flag


def complement(strand):
    complement_strand = ""
    n_dictionary = {"A" : "T", "T" : "A", "G" : "C", "C" : "G"}
    for x in strand:
        for base, complementary_base in n_dictionary.items():
            if x == base:
                complement_strand += complementary_base
    return complement_strand


dna = input("enter strand: ").strip().upper()
print("DNA strand: ", dna, end="\n")

if valid_strand(dna) == True:
    print("complementary strand: ", complement(dna))
else:
    print("the given strand is invalid")
```

```python
# create mRNA
# template strand: TACGCATGC
def valid_strand(template_strand):
    flag = False
    for x in template_strand:
            if x not in ["A", "T", "G", "C"]:
                flag = True
    return not flag 


def complement(template_strand):
    coding_strand = ""
    n_dictionary = {"A" : "T", "T" : "A", "G" : "C", "C" : "G"}
    for x in template_strand:
        for base, complementary_base in n_dictionary.items():
            if x == base:
                coding_strand += complementary_base
    return coding_strand


def transcribe(coding_strand):
    return coding_strand.replace("T", "U")


template_strand = input("enter template strand: ").strip().upper()
print("template strand:: ", template_strand, end="\n")

if valid_strand(template_strand) == True:
    coding_strand = complement(template_strand)
    print("coding strand: ", coding_strand, end="\n")
    mrna = transcribe(coding_strand)
    print("mRNA: ", mrna, end="\n")
else:
    print("template strand invalid")
```

```python
# translation
# template strand: TACGCATGC
aa_codon_dictionary = {
    "A": ["GCU", "GCC", "GCA", "GCG"],
    "R": ["CGU", "CGC", "CGA", "CGG", "AGA", "AGG"],
    "N": ["AAU", "AAC"],
    "D": ["GAU", "GAC"],
    "B": ["AAU", "AAC", "GAU", "GAC"],
    "C": ["UGU", "UGC"],
    "Q": ["CAA", "CAG"],
    "E": ["GAA", "GAG"],
    "Z": ["CAA", "CAG", "GAA", "GAG"],
    "G": ["GGU", "GGC", "GGA", "GGG"],
    "H": ["CAU", "CAC"],
    "I": ["AUU", "AUC", "AUA"],
    "L": ["CUU", "CUC", "CUA", "CUG", "UUA", "UUG"],
    "M": ["AUG"], 
    "K": ["AAA", "AAG"], 
    "F": ["UUU", "UUC"],
    "P": ["CCU", "CCC", "CCA", "CCG"],
    "S": ["UCU", "UCC", "UCA", "UCG", "AGU", "AGC"],
    "T": ["ACU", "ACC", "ACA", "ACG"],
    "W": ["UGG"],
    "Y": ["UAU", "UAC"],
    "V": ["GUU", "GUC", "GUA", "GUG"],
    "--STOP--": ["UAA", "UGA", "UAG"]
}


def valid_strand(template_strand):
    flag = False
    for x in template_strand:
        if x not in ["A", "T", "G", "C"]:
            flag = True
    return not flag
    

def complement(template_strand):
    coding_strand = ""
    n_dictionary = {"A" : "T", "T" : "A", "G" : "C", "C" : "G"}
    for x in template_strand:
            for base, complementary_base in n_dictionary.items():
                if x == base:
                    coding_strand += complementary_base
                    break
    return coding_strand


def transcribe(coding_strand):
    return coding_strand.replace("T", "U")


def check_mRNA_length(mrna):
    return len(mrna) >= 3


def translate(mrna):
    protein = ""
    
    for i in range(0, len(mrna) - 2, 3):
        codon = mrna[i:i+3]
        for aa, codons in aa_codon_dictionary.items():
            if codon in codons:
                if aa == "--STOP--":
                    return protein
                protein += aa
                break
    return protein



template_strand = input("enter template strand: ").strip().upper()

if valid_strand(template_strand) == True:
    print("template strand: ", template_strand, end="\n")
    coding_strand = complement(template_strand)
    print("coding strand: ", coding_strand, end="\n")
    mrna = transcribe(coding_strand)
    print("mRNA: ", mrna, end="\n")
    
    if check_mRNA_length(mrna) == True:
        protein = translate(mrna)
        print("protein: ", protein)
    else:
        print("length of mRNA too short to be translated.")
else:
    print("invalid DNA template strand.")
```

```python
# reverse and complement
# template strand: TACGCATGC
def valid_strand(template_strand):
    flag = False
    for x in template_strand:
        if x not in ["A", "T", "G", "C"]:
            flag = True
            break
    return not flag


def complement(template_strand):
    coding_strand = ""
    n_dictionary = {"A" : "T", "T" : "A", "G" : "C", "C" : "G"}
    for x in template_strand:
        for base, complementary_base in n_dictionary.items():
            if x == base:
                coding_strand += complementary_base
    return coding_strand


def reverse(strand):
    return strand[::-1]


template_strand = input("Enter the strand: ").strip().upper()

if valid_strand(template_strand) == True:
    print("template strand: ", template_strand, "\treverse: ", reverse(template_strand), end="\n")
    coding_strand = complement(template_strand)
    print("coding strand: ", coding_strand, "\treverse: ", reverse(coding_strand), end="\n")
else:
    print("input strand invalid")
```

```python
# gc content and at_gc fraction
def valid_strand(strand):
    flag = False
    for x in strand:
        if x not in ["A", "T", "G", "C"]:
            flag = True
            break
    return not flag


def gc_fraction(strand):
    count = 0
    for x in strand:
        if x in ["G", "C"]:
            count += 1
    return round(count / len(strand), 2)


def at_gc_fraction(strand):
    count_1, count_2 = 0, 0
    
    for x in strand:
        if x in ["C", "G"]:
            count_1 += 1
        else:
            count_2 += 1
    return round(count_2/count_1, 2)


dna = input("enter a strand: ").strip().upper()

if valid_strand(dna) == True:
    print("DNA strand: ", dna, end="\n")
    print("GC content: ", gc_fraction(dna), end="\n") 
    print("AT-CG fraction: ", at_gc_fraction(dna), end="\n")

else:
    print("input strand invalid")
```

```python
# reverse transcription
# mrna: AUGCGUACG
def valid_strand(mrna):
    flag = False
    for x in mrna:
        if x not in ["A", "U", "G", "C"]:
            flag = True
    return not flag


def complement(strand):
    complementary_strand = ""
    n_dictionary = {"A" : "T", "T" : "A", "G" : "C", "C" : "G"}
    for x in strand:
        for base, complement_base in n_dictionary.items():
            if x == base:
                complementary_strand += complement_base
                break
    return complementary_strand


def reverse_transcribe(mrna):
    sense_strand = mrna.replace("U", "T")
    return {"complementary strand" : complement(sense_strand), "sense strand: " : sense_strand}


mrna = input("enter mRNA strand: ").strip().upper()
if valid_strand(mrna) == True:
    print("mRNA: ", mrna, end="\n")
    print("DNA: ", reverse_transcribe(mrna), end="\n")
else:
    print("input strand invalid", end = "\n")
```
